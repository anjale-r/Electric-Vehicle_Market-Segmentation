The smallest number of components that explain around 90% of the variance is 9 principal components.

```python
from sklearn.cluster import KMeans


# Apply PCA to reduce to first 9 components
pca = PCA(n_components=9)
new_pca = pca.fit_transform(scaled_data)  # use scaled_data here

```

```python
# Elbow method with first 9 principal components
wcss = []
K_range = range(1, 11)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(new_pca)
    wcss.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(8, 5))
plt.plot(K_range, wcss, marker='o')
plt.title('The Elbow Method (Using First 9 Principal Components)')
plt.xlabel('Number of Clusters')
plt.ylabel('Within-cluster Sum of Squares (WCSS)')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score

kmeans = KMeans(n_clusters=8, random_state=42, n_init=10)
labels = kmeans.fit_predict(new_pca)  # using your PCA data

score = silhouette_score(new_pca, labels)
print(f'Silhouette Score for 2 clusters: {score:.3f}')
```
