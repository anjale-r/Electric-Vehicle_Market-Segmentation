#### Since the dataset contains both categorical and numerical variables, customer segmentation can be performed using two approaches:

#### 1. K-Means Clustering (after encoding):
This method involves converting categorical variables into numerical format using techniques such as one-hot encoding. Then, standard clustering algorithms like K-Means can be applied to the transformed data.

#### 2. K-Prototypes Clustering:
K-Prototypes is specifically designed to handle mixed-type data (both numeric and categorical). It combines the principles of K-Means (for numeric variables using Euclidean distance) and K-Modes (for categorical variables using matching dissimilarity).

###  K-Means Clustering

```python

from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.cluster import KMeans
```

```python
# Define columns
categorical_cols = ['Profession', 'Marrital Status', 'Education', 'Personal loan']
numeric_cols = ['Age', 'No of Dependents', 'Total Salary', 'Price']
```

```python
# Preprocess: One-hot encode + scale
preprocessor = ColumnTransformer([
    ('num', StandardScaler(), numeric_cols),
    ('cat', OneHotEncoder(), categorical_cols)
])

```

```python
X = preprocessor.fit_transform(data)
```

```python
wcss = []
K = range(1, 11)  # Trying k from 1 to 10

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)  # X is your preprocessed feature matrix
    wcss.append(kmeans.inertia_)

# Plotting the scree plot
plt.figure(figsize=(8, 5))
plt.plot(K, wcss, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia (WCSS)')
plt.title('Scree Plot / Elbow Method for K-Means')
plt.grid(True)
plt.show()



```

```python
from sklearn.metrics import silhouette_score

kmeans = KMeans(n_clusters=3,random_state=42, n_init=10)
labels = kmeans.fit_predict(X)  # using your PCA data

score = silhouette_score(X, labels)
print(f'Silhouette Score for 3 clusters: {score:.3f}')
```

```python

# Fit KMeans
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
data['KMeans_Cluster'] = kmeans.fit_predict(X)

# Cluster summary
summary = data.groupby('KMeans_Cluster').agg({
    'Age': 'mean',
    'Total Salary': 'mean',
    'Price': 'mean',
    'Profession': lambda x: x.mode()[0],
    'Education': lambda x: x.mode()[0]
})
print(summary)
```

```python


# Choose two meaningful numerical features
x_col = 'Total Salary'
y_col = 'Price'

plt.figure(figsize=(8, 6))
for cluster in data['KMeans_Cluster'].unique():
    subset = data[data['KMeans_Cluster'] == cluster]
    plt.scatter(subset[x_col], subset[y_col], label=f'Cluster {cluster}')

plt.xlabel(x_col)
plt.ylabel(y_col)
plt.title('K-Means Clusters')
plt.legend()
plt.grid(True)
plt.show()
```
